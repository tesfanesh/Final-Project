```python
# Name of the CSV file
Game='final-project/Game_Passing_df.csv'
season='final-project/new_df.csv'
```

 Reading Data in Pandas 

```python
# read csv file
df= pd.read_csv(Game)
df.head()
```

```python
# read csv file
df_new= pd.read_csv(season)
df_new.head()
```

```python
pd.merge(df_new, df, on='GameID')
```

```python
# filtering specific columns
Data_df = df.reindex(columns=["GameID", "Team", "Attempts", "Completions", "Drives", "Total_Yards","season"])
# Print df
Data_df.head()
```

```python
# save update data
Data_df.to_csv('merge.csv')
Data_df.head()
```
